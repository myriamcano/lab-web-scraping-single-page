```python
from bs4 import BeautifulSoup
import requests

import pandas as pd
```

```python
url = 'https://www.popvortex.com/music/charts/top-100-songs.php'
```

```python
response = requests.get(url)
response.status_code # 200 status code means OK!
```

```python
# 4.1. parse html (create the 'soup')
soup = BeautifulSoup(response.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
chart = soup.find_all('div', class_='feed-item music-chart flex-row')
```

```python
chart
```

```python

chart_positions = []
titles = []
artists = []
genres = []

for item in soup.select('.feed-item.music-chart.flex-row, .feed-item.music-chart.flex-row.new-release'):
    
    chart_position_element = item.find('p', class_='chart-position')
    chart_position_text = chart_position_element.text if chart_position_element else 'N/A'
    chart_positions.append({'Chart Position': chart_position_text})
    
    title_element = item.find('cite', class_='title')
    title_text = title_element.text if title_element else 'N/A'
    titles.append({'title': title_text})
    
    artist_element = item.find('em', class_='artist')
    artist_text = artist_element.text if artist_element else 'N/A'
    artists.append({'artist': artist_text})
    
    genre_element = item.find('a')
    genre_text = genre_element.text if genre_element else 'N/A'
    genres.append({'genre': genre_text})

# Create DataFrames from the lists
chart_positions_df = pd.DataFrame(chart_positions)
titles_df = pd.DataFrame(titles)
artists_df = pd.DataFrame(artists)
genres_df = pd.DataFrame(genres)

# Merge the DataFrames on their index
chart = pd.concat([chart_positions_df, titles_df, artists_df, genres_df], axis=1)
```
